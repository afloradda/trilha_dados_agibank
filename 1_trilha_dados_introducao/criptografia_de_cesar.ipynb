# Conceito: Cifra de C√©sar
Esta fun√ß√£o implementa a Cifra de C√©sar, um dos m√©todos de criptografia mais antigos da hist√≥ria. O algoritmo desloca cada letra do alfabeto por um n√∫mero fixo de posi√ß√µes.

Neste exerc√≠cio utilizaremos m√∫ltiplos conceitos de progrma√ß√£o para recordar a sintaxe Python. Segue a lista destes e onde os encontramos dentro dos blocos de c√≥digo:

- Estruturas de Repeti√ß√£o: for para iterar sobre cada caractere

- Estruturas Condicionais: if/else para tratar letras vs. outros caracteres

- Manipula√ß√£o de Listas: index() e acesso por posi√ß√£o

- Opera√ß√µes Matem√°ticas: M√≥dulo % 

- Fun√ß√µes: Modulariza√ß√£o e reutiliza√ß√£o de c√≥digo

- Par√¢metros: Passagem de dados entre fun√ß√µes

```python
def criptografar(texto, chave):
    ALFABETO = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    texto_criptografado = ""

    for letra in texto:
        if letra in ALFABETO:
            novo_index = (ALFABETO.index(letra) + chave) % 26
            texto_criptografado = texto_criptografado + ALFABETO[novo_index]
        else:
            texto_criptografado = texto_criptografado + letra
    
    return texto_criptografado
```

üí° Dica: O operador % garante que o resultado sempre fique entre 0-25, fazendo com que o √≠ndice "volte ao in√≠cio" quando ultrapassar 'z'.

```python
criptografar("heloisa", 5)
```

# Conceito: Processo Inverso
Para descriptografar, fazemos o caminho contr√°rio: subtra√≠mos a chave ao inv√©s de somar.

```python
def desincriptografar(variavel, chave):
    ALFABETO = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    # texto_criptografado = variavel
    texto_descriptografado = ""

    for letra in variavel:
        if letra in ALFABETO:
            index_origem = (ALFABETO.index(letra) - chave) % 26
            texto_descriptografado = texto_descriptografado + ALFABETO[index_origem]
        else:
            texto_descriptografado = texto_descriptografado + letra
    
    return texto_descriptografado
```

# Valida√ß√£o do Algoritmo
Este bloco demonstra o ciclo completo de criptografia e descriptografia:

```python
palavra = criptografar("heloisa", 5)

print(palavra)
```

```python
resultado = desincriptografar(palavra, 5)

print(resultado)
```
