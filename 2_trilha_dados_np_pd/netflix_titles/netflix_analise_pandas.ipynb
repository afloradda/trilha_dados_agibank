# Bloco 4 - Filtragem de Dados
**Filtros Booleanos B√°sicos**

# Bloco 8 - Cria√ß√£o de Dataset Espec√≠fico
**Filtragem Avan√ßada com Query**

```python
filmes_recentes = netflix_movies.query('type == "Movie" & release_year >= 2020')
filmes_recentes.head()
```

# Bloco 9 - Tratamento de Valores Ausentes
**Limpeza e Padroniza√ß√£o**

```python
filmes_recentes['country'] = filmes_recentes['country'].fillna("N√£o Registrado")
```

‚ö†Ô∏è Ponto de Aten√ß√£o: O DataFrame filmes_recentes foi criado atrav√©s de uma opera√ß√£o de filtragem. Sendo assim, o Pandas n√£o ter certeza se `filmes_recentes` √©: 

- Uma c√≥pia independente do DataFrame original

- Uma visualiza√ß√£o (view) que ainda est√° conectada ao DataFrame `netflix_movies`

```python
filmes_recentes.head()
```

Estrat√©gias de limpeza:

- fillna(): Substitui valores nulos (NaN) por valor padr√£o

- Verifica√ß√£o visual: head() para confirmar altera√ß√µes

- Padroniza√ß√£o: "N√£o Registrado" para dados faltantes

- Modifica√ß√£o in-place: Altera o DataFrame original

```python
# ‚úÖ C√ìDIGO CORRIGIDO (CLAUDE-SONNET-4) E SEGURO:
# 1. Criar filtro
filmes_recentes = netflix_movies.query('type == "Movie" & release_year >= 2020').copy()

# 2. Aplicar transforma√ß√µes com seguran√ßa
filmes_recentes['country'] = filmes_recentes['country'].fillna("N√£o Registrado")

# 3. Verificar resultado
print("‚úÖ Dados tratados com sucesso!")
filmes_recentes.head()
```
